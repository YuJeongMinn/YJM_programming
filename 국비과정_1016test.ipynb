# <지시사항 2>
## 1. age가 17을 초과하는지 여부를 True, False로 표현하는 legal_drinker열을 생성하시오.

```python
student["legal_drinker"] = student["age"].map(lambda x : True if x > 17 else False)
```

```python
student["legal_drinker"]
```
